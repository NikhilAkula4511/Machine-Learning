```python
y=df.pop('quality')
y
```

```python
X=df
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=42)
```

```python
X_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_trans = pd.DataFrame(scaler.fit_transform(X_train),columns=scaler.get_feature_names_out(),index=X_train.index)
```

```python
X_train_trans.describe()
```

```python
X_test_trans = pd.DataFrame(scaler.transform(X_test),columns=scaler.get_feature_names_out(),index = X_test.index)
X_test_trans
```

```python
X_test_trans.describe()
```

```python
y_test.unique()
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train_trans,y_train)
y_test_pred = reg.predict(X_test_trans)
y_test_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
acc_lr = accuracy_score(y_test,y_test_pred)
acc_lr
```

```python
print(classification_report(y_test,y_test_pred))
```

```python
print(confusion_matrix(y_test,y_test_pred))
```

```python
cm = confusion_matrix(y_test, y_test_pred)
cm_plot = sns.heatmap(cm, annot=True, cmap='Blues');
cm_plot.set_xlabel('Predicted Values')
cm_plot.set_ylabel('Actual Values')
```

```python
from sklearn.ensemble import RandomForestClassifier
reg = RandomForestClassifier()
reg.fit(X_train_trans,y_train)
print('\n<---------------- Predicted values  by model ------------------->\n')
y_test_pred = reg.predict(X_test_trans)
print(y_test_pred)
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print('\n<---------------- Accuracy Score ------------------->\n')
acc_rf=accuracy_score(y_test,y_test_pred)
print(acc_rf)
print('\n<---------------- Classification Report ------------------->\n')
print(classification_report(y_test,y_test_pred))
cm = confusion_matrix(y_test, y_test_pred)
print('\n<---------------- Confusion Matrix ------------------->\n')
print(cm)
print('\n<---------------- Correlation Map to Confusion Matrix ------------------->\n')
cm_plot = sns.heatmap(cm, annot=True, cmap='Blues');
cm_plot.set_xlabel('Predicted Values')
cm_plot.set_ylabel('Actual Values')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
reg = DecisionTreeClassifier()
reg.fit(X_train_trans,y_train)
print('\n<---------------- Predicted values  by model ------------------->\n')
y_test_pred = reg.predict(X_test_trans)
print(y_test_pred)
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print('\n<---------------- Accuracy Score ------------------->\n')
acc_DTC = accuracy_score(y_test,y_test_pred)
print(acc_DTC)
print('\n<---------------- Classification Report ------------------->\n')
print(classification_report(y_test,y_test_pred))
cm = confusion_matrix(y_test, y_test_pred)
print('\n<---------------- Confusion Matrix ------------------->\n')
print(cm)
print('\n<---------------- Correlation Map to Confusion Matrix ------------------->\n')
cm_plot = sns.heatmap(cm, annot=True, cmap='Blues');
cm_plot.set_xlabel('Predicted Values')
cm_plot.set_ylabel('Actual Values')
plt.show()
```

```python
#------------>Accuracy Comparison<------------
models = ['LR','RF','DTC']
acc_scores = [acc_lr,acc_rf,acc_DTC]
print("Models\tAccuracy\n")
for i in range(3):
    print(str(models[i]) + "\t" + str(acc_scores[i]),end='\n')
```

```python
#------------>Finding the Best Algorithm<----------

print("\n------------>We decide from the Bar Graph that the best Model<------------")
plt.rcParams['figure.figsize'] = (5,5)
plt.bar(models, acc_scores, color=['lightblue', 'pink', 'lightgreen'])
plt.ylabel("accuracy scores")
plt.title("Model Comparison based on Accuracy")
plt.show()
```
